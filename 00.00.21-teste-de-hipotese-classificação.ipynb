```python
import pandas as pd
import scipy as sp
import seaborn as sns
from scipy.stats import shapiro
import matplotlib.pyplot as plt
from scipy.stats import f_oneway
import numpy as np
from statsmodels.stats.multicomp import MultiComparison
```

```python
# Importando o dataset para uma variável
resultados = pd.read_csv('cls01-resultados_algo.csv')

# dropando as colunas que não serão utilizadas
resultados.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
# Teste de hipótese
# Shapiro-Wilk

alpha = 0.05
```

```python
# resultados da teste de Shapiro-Wilk

print('Resultados Arvores de decisão : ', shapiro(resultados['Decision Tree']))
print('Resultados Random Forest : ', shapiro(resultados['Random Forest']))
print('Resultados K-Nearest Neighbors : ', shapiro(resultados['K-Nearest Neighbors']))
print('Resultados Logistic Regression : ', shapiro(resultados['Logistic Regression']))
print('Resultados SVM : ', shapiro(resultados['SVM']))
print('Resultados Neural Network : ', shapiro(resultados['Neural Network']))
```

```python
# teste de hipótese
# ANOVA ( análise de variação )
# comparação entre a variaçãoo dos valores de acurácia dos algoritmos
```

```python
p = f_oneway(resultados['Decision Tree'], 
             resultados['Random Forest'], 
             resultados['K-Nearest Neighbors'], 
             resultados['Logistic Regression'], 
             resultados['SVM'], 
             resultados['Neural Network'])
```

```python
p
```

```python
if p[1] <= alpha:
    print('Hipotese nula rejeitada. Dados são diferentes.')

else:
    print('Hipotese alternativa rejeitada. Não há diferença entre os algoritmos')
```
