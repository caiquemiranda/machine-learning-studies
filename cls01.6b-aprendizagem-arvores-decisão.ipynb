```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import pickle as pkl
import matplotlib.pyplot as plt
```

```python
with open('./data/risc_credit.pkl', mode='rb') as f:
    X_risc_credit, y_risc_credit = pkl.load(f)   
```

```python
dt_risc_credit = DecisionTreeClassifier(criterion='entropy')
dt_risc_credit.fit(X_risc_credit, y_risc_credit)
```

```python
dt_risc_credit.feature_importances_
```

```python
from sklearn.tree import export_graphviz
from sklearn import tree
```

```python
tree.export_graphviz(dt_risc_credit, out_file='tree.dot')
```

```python
tree.plot_tree(dt_risc_credit);
```

```python
fig, axis = plt.subplots(1, 1, figsize=(20, 10))
previsoes = ['historia', 'divida', 'garantias', 'renda']
tree.plot_tree(dt_risc_credit, feature_names=previsoes, filled=True); #  class_names=X_risc_credit.classes_, retorna os nomes das classes
```
