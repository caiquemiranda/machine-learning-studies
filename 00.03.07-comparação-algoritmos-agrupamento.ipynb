```python
# importação das bibliotecas necessarias

import plotly.express as px                             # geração de graficos
from sklearn import datasets                            # geração de datasets
from sklearn.cluster import KMeans                      # algoritmo K-means
from sklearn.cluster import AgglomerativeClustering     # algoritmo de Agrupamento hierarquico
from sklearn.cluster import DBSCAN                      # algortimo BDscan
```

```python
# criando um dataset com 1_500 registros

X_random, y_random = datasets.make_moons(n_samples=1_500, noise=0.09)
```

```python
# visualização grafica dos dados

grafico = px.scatter(x=X_random[:, 0], y=X_random[:, 1])
grafico.show()
```

```python
# aplicando os algoritmos de agrupamento para comparação
```

```python
# Kmeans

kmeans = KMeans(n_clusters=2)
rotulos = kmeans.fit_predict(X_random)
grafico = px.scatter(x = X_random[:,0], y = X_random[:, 1], color = rotulos)
grafico.show()
```

```python
# agrupamento hierarquico

hc = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
rotulos = hc.fit_predict(X_random)
grafico = px.scatter(x = X_random[:,0], y = X_random[:, 1], color = rotulos)
grafico.show()
```

```python
# BDscan

dbscan = DBSCAN(eps=0.1)
rotulos = dbscan.fit_predict(X_random)
grafico = px.scatter(x = X_random[:,0], y = X_random[:, 1], color = rotulos)
grafico.show()
```
