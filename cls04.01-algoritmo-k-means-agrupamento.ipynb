```python
# importando as bibliotecas necessarias

import plotly.express as px                       # visualização de graficos
import plotly.graph_objects as go                 # visualização de graficos
import numpy as np                                # calculos matematicos
from sklearn.preprocessing import StandardScaler  # padronização dos dados
from sklearn.cluster import KMeans                # algortimo de agruoamento K-means
from sklearn.datasets import make_blobs           # geração de dados aleatórios
```

```python
# trabalhando com dados randomicos
```

```python
# gerando valores randomicos 

X_random, y_random = make_blobs(n_samples=300, 
                                centers=5, 
                                random_state=1)
```

```python
# visualizando dados randomicos (X_random)

X_random
```

```python
# visualizando dados gerados (y_random)

y_random
```

```python
# visualizando dados graficamente

grafico = px.scatter(x=X_random[ :, 0], y=X_random[ :, 1])
grafico.show()
```

```python
# definindo modelo

kmeans_blobs = KMeans(n_clusters=5)
kmeans_blobs.fit(X_random)
```

```python
# visualizando registros em seus grupos

rotulos = kmeans_blobs.predict(X_random)
rotulos
```

```python
centroides = kmeans_blobs.cluster_centers_
centroides
```

```python
# visualização grafica dos agrupamentos

grafico1 = px.scatter(x= X_random[ :, 0], y= X_random[ :, 1], color=rotulos)
grafico2 = px.scatter(x= centroides[ :, 0], y= centroides[ :, 1], size=[5, 5, 5,5,5])
grafico3 = go.Figure(data= grafico1.data + grafico2.data)

grafico3.show()
```
