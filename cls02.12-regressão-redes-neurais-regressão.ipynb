```python
# selecionando as variaveis para o modelo
# tamanho da casa - sqft_living

X_base_casas = base_casas.iloc[ :, 3:19].values
X_base_casas
```

```python
# seleção da variavel para o modelo
# preço da casa - price [target]

y_base_casas = base_casas.iloc[ :, 2].values
y_base_casas
```

```python
# divisão da base de dados em treino e teste

X_casas_treinamento, X_casas_teste, y_casas_treinamento, y_casas_teste = train_test_split(X_base_casas, y_base_casas, test_size=0.3, random_state=0)
```

```python
# tamnho dos dados de treino e teste

X_casas_treinamento.shape, X_casas_teste.shape, y_casas_treinamento.shape, y_casas_teste.shape
```

```python
# normalização dos dados de treinamento e teste

scaler_x_casas = StandardScaler()
X_casas_treinamento_scaled = scaler_x_casas.fit_transform(X_casas_treinamento)

scaler_y_casas = StandardScaler()
y_casas_treinamento_scaled = scaler_y_casas.fit_transform(y_casas_treinamento.reshape(-1, 1))
```

```python
# visualização do tamanho dos dados de treino

X_casas_treinamento_scaled.shape, y_casas_treinamento_scaled.shape
```

```python
# normalização dos dados de teste

X_casas_teste_scaled = scaler_x_casas.transform(X_casas_teste)
y_casas_teste_scaled = scaler_y_casas.transform(y_casas_teste.reshape(-1, 1))
```

```python
# visualização do tamanho dos dados de teste

X_casas_teste_scaled.shape, y_casas_teste_scaled.shape
```

```python
# criação do modelo e treinamento
# 16 neurônios entrada => 9 neurônios na camada oculta => 9 neurônios na camada oculta => 1 neurônio na camada de saída

regressor_casas_nn = MLPRegressor(max_iter=1000, hidden_layer_sizes=(9, 9))
regressor_casas_nn.fit(X_casas_treinamento_scaled, y_casas_treinamento_scaled.ravel())
```

```python
# score do modelo na base de dados de treino

regressor_casas_nn.score(X_casas_treinamento_scaled, y_casas_treinamento_scaled)
```

```python
# score do modelo na base de dados de teste

regressor_casas_nn.score(X_casas_teste_scaled, y_casas_teste_scaled)
```

```python
# previsão dos dados de teste

previsoes_casas = regressor_casas_nn.predict(X_casas_teste_scaled)
previsoes_casas
```

```python
# dados reais

y_casas_teste_scaled
```

```python
# desescalonamento dos dados 

y_casas_teste_inversed = scaler_y_casas.inverse_transform(y_casas_teste_scaled)
previsoes_casas_inversed = scaler_y_casas.inverse_transform(previsoes_casas.reshape(-1, 1))
```

```python
# dados reais

y_casas_teste_inversed
```

```python
# dados previstos

previsoes_casas_inversed
```

```python
# erro médio quadrático

mean_absolute_error(y_casas_teste_inversed, previsoes_casas_inversed)
```
